```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Part 1
b = 1
w1 = 3
x = 0

y = (w1 * x) + b

print(f"The linear regression prediction is: {y}")

x_vals = np.linspace(-10, 10, 100)
y_vals = w1 * x_vals + b

plt.axhline(0, color = 'black')
plt.axvline(0, color = 'black')
plt.plot(x_vals, y_vals)
plt.title("Linear Regression Prediction")
plt.show()
```

## **Analysis**
This graph is a visual representation of the linear regression model y = 3x + b. When the x = 0, it predicts the value of one. This shows us that the slope is what controls how much the output increases and decreases alongside the input (x).

```python
# Part 2

b = 1
w1 = 3
w2 = -2
w3 = 4

# Example 1
x1a = 3
x2a = -1
x3a = 2

y1 = (w1 * x1a) + (w2 * x2a) + (w3 * x3a) + b

# Example 2
x1b = 10
x2b = 0
x3b = -10

y2 = (w1 * x1b) + (w2 * x2b) + (w3 * x3b) + b

print(f"The linear regression prediction for example 1 is {y1}.")
print(f"The linear regression prediction for example 2 is {y2}.")

contrib1 = [w1*x1a, w2*x2a, w3*x3a, b]
contrib2 = [w1*x1b, w2*x2b, w3*x3b, b]

labels = ['w1 * x1', 'w2 * x2', 'w3 * x3', 'bias']
x = np.arange(len(labels))

plt.bar(x - 0.2, contrib1, width=0.4, label='Example 1')
plt.bar(x + 0.2, contrib2, width=0.4, label='Example 2')

plt.axhline(0, color='black')
plt.axvline(-0.5, color='black')

plt.xticks(x, labels)
plt.title('Linear Regression Prediction (Part 2)')
plt.legend()

plt.show()
```

## **Analysis**
The linear regression model predicts the value of 20 for the first example. This is due to the values being all positive, giving a high number of output. On the otherhand, the second example has a low output of -9 due x3 having a large negative value. The graph shows us just how greatly the components affect the overall output. The blue bars represent example 1 while the other represents example 2. With this visualization, we can easily determine why the output is high or low.
