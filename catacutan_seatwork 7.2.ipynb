### Exercise Part 4:

Loading the csv file

```python
import pandas as pd
df = pd.read_csv("/content/Meteorite_Landings.csv")
```

Preparing the data

```python
df["year"] = df["year"].str.slice(6,10)
df["year"] = pd.to_numeric(df["year"])

df["mass (g)"] = pd.to_numeric(df["mass (g)"])

df_filtered = df[(df["year"] >= 2005) & (df["year"] <= 2009)]
```

Create a pivot table

```python
pivot = pd.pivot_table(
    df_filtered,
    index = "year",
    columns = "fall",
    values = "mass (g)",
    aggfunc = {"mass (g)": ["count",("95th_Percentile", lambda x: x.quantile(0.95))]}
)

pivot.columns = ["_".join(col).replace("<lambda>", "95th_Percentile") for col in pivot.columns]


print(pivot)
```

Compare the summary of statistics

```python
summary = df.groupby("fall")["mass (g)"].describe()
print(summary)
```

### Exercise Part 5:

Loading the csv and preparing the data

```python
df_taxi = pd.read_csv("/content/2019_Yellow_Taxi_Trip_Data.csv")
df_taxi["tpep_dropoff_datetime"] = pd.to_datetime(df_taxi["tpep_pickup_datetime"])

df_taxi = df_taxi.set_index("tpep_dropoff_datetime")
```

Resampling the data and calculating the total trip_distance, fare_amount, tolls_amount, and tip_amount.

```python
hourly = df_taxi.resample('H').agg({
    'trip_distance': 'sum',
    'fare_amount': 'sum',
    'tolls_amount': 'sum',
    'tip_amount': 'sum'
})

print(hourly.head())
```

Finding the 5 hours with the most tips

```python
most_tips = hourly.sort_values('tip_amount', ascending=False).head(5)
print(most_tips)
```
