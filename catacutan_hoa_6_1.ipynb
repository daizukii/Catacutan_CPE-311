##Exercise 1
Run the given code below for exercises 1 and 2, perform the given tasks without using any Python modules.

```python
import random
random.seed(0)
salaries = [round(random.random()*1000000, -3) for _ in range(100)]
```

Using the data generated above, calculate the following statistics without importing anything from the statistics module in the standard library (https://docs.python.org/3/library/statistics.html) and then confirm your results match up to those that are obtained when using the statistics module (where possible):

* Mean
* Median
* Mode (hint: check out the Counter in the collections module of the standard library at https://docs.python.org/3/library/collections.html#collections.Counter)
* Sample variance
* Sample standard deviation

```python
#without importing from the statistics module
mean = sum(salaries)/len(salaries)

if len(salaries)%2 == 1:
  median = sorted(salaries)[len(salaries)//2]
else:
  median = (sorted(salaries)[len(salaries)//2-1] + sorted(salaries)[len(salaries)//2-1])/2

mode = max(set(salaries), key=salaries.count)
sample_variance = sum((x - mean)**2 for x in salaries)/(len(salaries)-1)
sample_sd = sample_variance**0.5

print(f"The mean of the salaries is {mean}.")
print(f"The median of the salaries is {median}.")
print(f"The mode of the salaries is {mode}.")
print(f"The sample variance of the salaries is {sample_variance}.")
print(f"The sample standard deviation of the salaries is {sample_sd}.")
```

```python
#importing the statistics module

from statistics import mean, median, mode, variance, stdev

mean = mean(salaries)
median = median(salaries)
mode = mode(salaries)
sample_variance = variance(salaries)
sample_sd = stdev(salaries)

print(f"The mean of the salaries is {mean}.")
print(f"The median of the salaries is {median}.")
print(f"The mode of the salaries is {mode}.")
print(f"The sample variance of the salaries is {sample_variance}.")
print(f"The sample standard deviation of the salaries is {sample_sd}.")
```

##Exercise 2
Using the same data, calculate the following statistics using the functions in the statistics module where appropriate:
* Range
* Coefficient of variation Interquartile range
* Quartile coefficient of dispersion


```python
from statistics import quantiles

n = len(salaries)
range = max(salaries) - min(salaries)
coefficient_of_variation = sample_sd / mean
q1 = quantiles(salaries, n=4)[0]
q3 = quantiles(salaries, n=4)[2]
interquartile_range = q3 - q1
qcd = (q3 - q1)/(q3 + q1)

print(f"Range: {range}")
print(f"Coefficient of Variation: {coefficient_of_variation}")
print(f"Interquartile Range: {interquartile_range}")
print(f"Quartile Coefficient of Dispersion: {qcd}")
```
